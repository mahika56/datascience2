```python
import yfinance as yf
tesla = yf.Ticker("TSLA")
financials = tesla.financials
print("Tesla Financials:")
print(financials)
if 'Total Revenue' in financials.index:
    revenue_data = financials.loc['Total Revenue']
    print("\nTesla Revenue Data:")
    print(revenue_data.tail())  
else:
    print("Revenue data not found.")
revenue_data.to_csv('tesla_revenue_yahoo.csv', header=True)
```

```python
import yfinance as yf
gme = yf.Ticker("GME")
gme_data = gme.history(period="5y")
gme_data = gme_data.reset_index()
print(gme_data.head())
gme_data.to_csv('gme_stock_data.csv', index=False)
```

```python
import yfinance as yf
import pandas as pd
gme = yf.Ticker("GME")
income_statement = gme.quarterly_financials
if 'Total Revenue' in income_statement.index:
    gme_revenue = income_statement.loc['Total Revenue']
    gme_revenue = gme_revenue.reset_index()
    gme_revenue.columns = ['Date', 'Revenue']
    print(gme_revenue.tail())
    gme_revenue.to_csv('gme_revenue.csv', index=False)
else:
    print("Revenue data not found.")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
def make_graph(stock_data, title):
    plt.figure(figsize=(10, 5))
    plt.plot(stock_data['Close'], label='Tesla Stock Price', color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="1y") 
make_graph(tesla_data, "Tesla Stock Price - Last 1 Year")
```
